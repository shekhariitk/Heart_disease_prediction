```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
heart_df.drop(columns=['education'],inplace=True)
```

```python
heart_df['Gender'] =heart_df['Gender'].replace(['Male','Female'],[1,0])
```

```python
heart_df['prevalentStroke'].unique()
```

```python
heart_df['prevalentStroke'] = heart_df['prevalentStroke'].replace(['yes','no'],[1,0])
```

```python
heart_df['Heart_ stroke'] = heart_df['Heart_ stroke'].replace(['yes','No'],[1,0])
```

```python
heart_df.reset_index(inplace=True)
```

```python
heart_df.drop(columns=['index'],inplace=True)
```

```python
heart_df
```

```python
input_data = heart_df.drop(columns=['Heart_ stroke'])
output_data = heart_df['Heart_ stroke']
```

```python
output_data.value_counts()
```

```python
# DATA RESAMPLING
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sn = SMOTE(random_state=2)
```

```python
input_data , output_data = sn.fit_resample(input_data,output_data.ravel())
```

```python
np.unique(output_data,return_counts=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train , x_test ,y_train , y_test = train_test_split(input_data,output_data,test_size=0.2)
```

```python
x_test
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
predict
```

```python
model.score(x_test, y_test)
```
